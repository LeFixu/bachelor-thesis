```python
import pandas as pd

# read the data
female_df = pd.read_excel('female.xlsx', engine='openpyxl', index_col=None, header=0)
female_df.head(3)

female_df['firstname'] = female_df['firstname'].str.lower()
```

```python
female_df
```

```python
male_df = pd.read_excel('male.xlsx', engine='openpyxl', index_col=None, header=0)
male_df.head(3)

male_df['firstname'] = male_df['firstname'].str.lower()
```

```python
test_name = 'Elia' # female: 531, male: 4511

clean_name = test_name.lower()

loc_male = male_df.loc[(male_df['firstname'] == clean_name)]
loc_female = female_df.loc[(female_df['firstname'] == clean_name)]

if loc_male.any().all():
    print('name is male')
    print(loc_male['Total'].values[0])
if loc_female.any().all():
    print('name is female')
    print(loc_female['Total'].values[0])
```
