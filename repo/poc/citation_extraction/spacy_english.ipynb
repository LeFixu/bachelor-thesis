```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

print("text", "lemma_", "pos_", "tag_", "dep_", "shape_", "is_alpha", "is_stop")
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

# en_core_web_md

```python
!python -m spacy download en_core_web_md
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_md")

def print_verbs(doc):
    print("text", "lemma_", "pos_", "tag_", "dep_", "shape_", "is_alpha", "is_stop")
    for token in doc:
        if token.pos_ != "VERB":
            continue
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
                token.shape_, token.is_alpha, token.is_stop)

def print_sentence(sentence):
    doc = nlp(sentence)

    print(doc.text)
    print()
    displacy.render(doc, style="dep", options={"compact":True})
    print_verbs(doc)
```

```python
import spacy
from nltk import Tree

def to_nltk_tree(sentence):
    def tok_format(tok):
        return "_".join([tok.orth_, tok.dep_])
    def to_nltk_tree_(node):
        if node.n_lefts + node.n_rights > 0:
            return Tree(tok_format(node), [to_nltk_tree_(child) for child in node.children])
        else:
            return tok_format(node)
        
    doc = nlp(sentence)
    return [to_nltk_tree_(sent.root).pretty_print() for sent in doc.sents]

to_nltk_tree("The quick brown fox jumps over the lazy dog.")
```

```python
sentence = 'In industries where the government wants to establish  champions, "companies are being disadvantaged,” he says.'
to_nltk_tree(sentence)
```

```python
sentence = '"These regulations represent a complete violation of all rules of the European single market,” complains one Heidelberg executive.'
to_nltk_tree(sentence)
```
